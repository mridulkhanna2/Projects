# 1. Log in Instagram

```python
def login(user,password):
    driver.get('https://www.instagram.com/?hl=en')
    time.sleep(5)
    assert 'Instagram' in driver.title
    us=driver.find_element_by_name("username")
    us.send_keys(user)
    ps=driver.find_element_by_name("password")
    ps.send_keys(password)
    bt=driver.find_element_by_xpath("//button[contains(@class,'y3zKF')]")
    bt.submit()
    time.sleep(5)
```

```python
#run this if you have set up two factor authentication

def verification(x):
    sc=driver.find_element_by_name("verificationCode")
    sc.send_keys(x)
    bt=driver.find_element_by_xpath("//button[contains(@class,'y3zKF')]")
    bt.submit()
    time.sleep(5)
```

# 3. Open profile

Enter the username for this function to be applied on any Instagram handle

```python
#To open any instagram profile or handle- enter the username in place of handle
def open_profile(handle):
    driver.get('https://www.instagram.com/?hl=en')
    time.sleep(5)
    search=driver.find_element_by_xpath("//div[contains(@class,'LWmhU _0aCwM')]/input")
    search.clear()
    search.send_keys(handle)
    time.sleep(5)
    
    so_del=driver.find_element_by_xpath("//div[@class='fuqBx']/a")
    so_del.click()
```

# 4. Follow/Unfollow :

This function can be implied on any Instagram handle.

```python
open_profile("So Delhi")
```

### Follow

```python
def follow(profile):
    
    open_profile(profile)
    try:
        follow=driver.find_element_by_xpath("//button[contains(@class,'_5f5mN       jIbKX  _6VtSN     yZn4P   ')]")
        if follow != 'Follow':
            follow.click()
            print("Followed")
    
    except NoSuchElementException:
        print('Already following')
```

```python
follow("So Delhi")
```

### Unfollow

```python
def unfollow(profile):
    open_profile(profile)
    time.sleep(4)
    
    try:
        unfollow=driver.find_element_by_xpath("//button[contains(@class,'-fzfL')]")
        unfollow.click()
        time.sleep(5)
     
    except NoSuchElementException:
        print('You are not following')
        
        
    try:
        unfollow=driver.find_element_by_xpath("//button[contains(@class,'Cab_   ')]")
        if unfollow.text == 'Unfollow':
            unfollow.click()
            print("Unfollowed")
            
    except NoSuchElementException:
        pass

        
    
```

```python
unfollow("So Delhi")
```

# 5. Like/Unlike posts:

These functions can be applied on any Instagram handle.
<br>(except private accounts not followed by you)

```python
def like_post(profile):
    open_profile(profile) # calling func to open given profile
    time.sleep(5)
    try:
        
        post=driver.find_element_by_class_name("_9AhH0")
        post.click()
        time.sleep(4)
        
    
        svg=driver.find_elements_by_class_name("_8-yf5 ") 


    #     so that this code can be applicable to different pages with variety of content
        if ((svg[9].get_attribute('aria-label') == 'Like') or (svg[9].get_attribute('aria-label') == 'Unlike')):
            x=9
        elif ((svg[10].get_attribute('aria-label') == 'Like') or (svg[10].get_attribute('aria-label') == 'Unlike')):
            x=10
        elif ((svg[11].get_attribute('aria-label') == 'Like') or (svg[11].get_attribute('aria-label') == 'Unlike')):
            x=11
        elif ((svg[12].get_attribute('aria-label') == 'Like') or (svg[12].get_attribute('aria-label') == 'Unlike')):
            x=12

        for i in range(30):
            time.sleep(10)
            driver.execute_script('window.scrollBy(0,100);')
            time.sleep(5)
            svg=driver.find_elements_by_class_name("_8-yf5 ")


            if svg[x].get_attribute('aria-label') != 'Unlike':
                like=driver.find_element_by_xpath("//section[contains(@class,'ltpMr Slqrh')]/span/button")
                like.click()
                time.sleep(5)
                print("Liked post no",i+1)
            else:
                print("Post no", i+1 ,"is already liked by you")


            next=driver.find_element_by_xpath("//a[contains(@class,' _65Bje  coreSpriteRightPaginationArrow')]")
            next.click()

            if (i% 6 == 0):
                driver.execute_script('window.scrollBy(0,500);') 
                time.sleep(5)


        close=driver.find_element_by_xpath("//div[contains(@class,'                    Igw0E     IwRSH      eGOV_         _4EzTm                                                                                  BI4qX            qJPeX            fm1AK   TxciK yiMZG')]/button")
        close.click()
        
#to handle private accounts and accounts < 30 posts
    except NoSuchElementException:
            pass
        
```

# 6. List of Followers

This function can be applied on any Instagram handle.
<br>(except private accounts not followed by you)

```python
def click(profile):
    
    open_profile(profile)
    time.sleep(5)
    a=driver.find_elements_by_class_name("Y8-fY ")
    a[1].click()
    time.sleep(4)
    followers500()
```

```python
def followers500():    
    for i in range(50):
        current_height= driver.execute_script('return document.getElementsByClassName("isgrP")[0].scrollHeight;')
        driver.execute_script('document.getElementsByClassName("isgrP")[0].scrollTo(0, arguments[0]);',current_height) 
        time.sleep(1)
    time.sleep(4)
    foll()
```

```python
def foll():
    foodtalk=[]
    we=driver.execute_script('return document.getElementsByClassName("FPmhX");')
    time.sleep(3)
    for i in we:
        foodtalk.append(i.text)
    foodtalk=foodtalk[:500]
    print("List of first 500 followers of foodtalkindia")
    count=0
    for i in foodtalk:
        print(i)
        count+=1
    print()
    print("Total followers: ",count) #just to make sure that every follower is printed
    time.sleep(2)
```

```python
click('foodtalkindia')
```

```python
click("sodelhi")
```

# 7. Checking status of story

This function can be applied on any Instagram handle.
<br>(except private accounts not followed by you)

```python
def story_status(profile):
    open_profile(profile)
    time.sleep(5)
    
    try:
        st=driver.find_element_by_xpath("//div[contains(@class,'h5uC0')]")
        
        h=int(driver.find_element_by_class_name("CfWVH").get_attribute('height'))
        #  Height of canvas helps to determine status of story
        if (h == 137 or h==252):
            st.click()
            print("Seeing the story right now")
        else:
            print("Story on this page is already seen by you")

    except NoSuchElementException:
        print("No story is uploaded on this page")
    
```

```python
story_status("coding.ninjas")
```
